```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from crewai import LLM
llm = LLM(
    model = "gemini/gemini-2.5-flash-lite",
    temperature=0.1
)
llm.call("How LLM work??.")
```

```python
# basic agent in crewai
from crewai import Agent , Task , Crew

email_assistant = Agent(
    role="Email assistant agent",
    goal="improve email and make them sound fully professional and clear",
    backstory="A highly experiance communication expert skilled in professional email writing",
    verbose = "True",
    llm=llm
)

original_email = """
I would like to inform you that I need leave for 3 days from 26 feb to 1 march to attend an important family function.

I will complete my assigned tasks before leaving and ensure a proper handover if required. I will also be available on call in case of any urgent matter.

Kindly approve my leave request."""

email_task = Task(
    description=f"""take the following email you can see this is
     very rufe email your task is that to make it sound like fully professional and abel to approve by HR team . Expand abbreviation '''{original_email}'''""",
     agent=email_assistant,
     expected_output= """a professional email with proper Email formating and content"""
)
crew = Crew(
    agents=[email_assistant],
    tasks=[email_task],
    verbose=True  # it will print more output
)

result = crew.kickoff()
print(result)
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from crewai import LLM
llm = LLM(
    model = "gemini/gemini-2.5-flash-lite",
    temperature=0.1
)
```

```python
original_email = """
looping in Priya. TAS and PRX updates are in the deck. ETA for SDS integration is Friday.
Let's sync up tomorrow if SYNCBOT allows ðŸ˜„. ping me if any blockers.."""

from crewai.tools import BaseTool

class ReplaceJargonsTool(BaseTool):
    name: str = "JargonReplacementTool"
    description : str = "Replaces internal abbreviations in the email with full forms."

    def _run(self, email: str) -> str:
        replacements = {
            "PRX": "Project Phoenix (internal AI revamp project)",
            "TAS": "technical architecture stack",
            "DBX": "client database cluster",
            "SDS": "Smart Data Syncer",
            "SYNCBOT": "internal standup assistant bot",
            "WIP": "in progress",
            "POC": "proof of concept",
            "ping": "reach out"
        }
        for jargon, replacement in replacements.items():
            email = email.replace(jargon, replacement)

        return email

    
jt = ReplaceJargonsTool()
jt.run(original_email)
```

```python
from crewai import Agent , Task , Crew

email_assistant = Agent(
    role="Professional Email Writing Assistant",
    goal="Expand abbreviations using the tool and rewrite email professionally.",
    backstory="Expert in corporate communication and clarity.",
    tools = [jt],
    verbose = "True",
    llm=llm
)


email_task = Task(
    description=f"""
First, use the JargonReplacementTool to expand all internal abbreviations
in the following email.

Then rewrite the expanded email in a fully professional format
with proper structure and clarity.

Email:
{original_email}
""",
    agent=email_assistant,
    expected_output="A professionally written email with expanded abbreviations."
)
crew = Crew(
    agents=[email_assistant],
    tasks=[email_task],
    verbose=True  # it will print more output
)

result = crew.kickoff()
print(result)
```
